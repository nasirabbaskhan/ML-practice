```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("insurance3.csv")
```

```python
df
```

```python
df["brought_insurance"].replace({"no":0,"yes":1},inplace=True)
```

```python
plt.scatter(x="age", y="brought_insurance", data=df)
```

```python
# split data into test and training ser

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[["age"]],df["brought_insurance"],test_size=0.2)
```

```python
# import logisticregressiom
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
# train the model by feeding x_train,y_train data
lr.fit(x_train,y_train)
```

```python
# predict the model

lr.predict(x_test)
```

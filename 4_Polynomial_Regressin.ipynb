```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("salary.csv")
```

```python
df  # dependent variable is salery and infependent variable is Level
```

```python
x = df.iloc[:,1:2].values
```

```python
x
```

```python
y = df.iloc[:,2].values
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
# plot best-fit line that determine which algorithem is best to solve the problem
sns.lmplot(x="Level",y="Salary", data=df) 
```

```python

# plynomial regression is also linear regression but have degree 2
# before apply linear regression convet independent variables x in x^2 
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2) #it will convert x in x^2 
```

```python
x_poly = poly.fit_transform(x) # now it has been converted x in x^2
```

```python
#train_test_split 
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test  = train_test_split(x_poly,y,test_size=0.2, random_state=42)
```

```python
# import LinearRegression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
# train the model
reg.fit(x_train,y_train)
```

```python
# predict  the model
y_pred = reg.predict(x_test)
```

```python
y_pred 
```

```python
reg.score(x_test, y_pred)
```

```python
# if you want to predict new data,before convert it into polinomial feature then predict it
# e.g. I want to predict the salary for level 11
```

```python
poly.transform([[11]])    # convertinf 11 and 7 into polinomial feature
```

```python
reg.predict(poly.transform([[7]]))
```

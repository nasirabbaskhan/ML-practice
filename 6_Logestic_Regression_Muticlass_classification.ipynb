```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("multi_regression.csv")
```

```python
df
```

```python
# converting species in binery

from sklearn.preprocessing import LabelEncoder
```

```python
df["Species"].unique()
```

```python
df["Species"].replace({'Ires-Virginica':"1", 'Ires-Setosa':"2", 'Ires-Versicolor':"3", 'Ites-Setosa':"4"},inplace=True)
```

```python
df
```

```python
# split into training and testing data sets
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[["SepelLengthCm","SepelWidthCm","PetalLengthCm","PetalWidthCm"]],df["Species"],test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
# train the model
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)  # {'Ires-Virginica':"1", 'Ires-Setosa':"2", 'Ires-Versicolor':"3", 'Ites-Setosa':"4"}
```

```python
x_test
```

```python
# check the model score

lr.score(x_test,y_test)
```
